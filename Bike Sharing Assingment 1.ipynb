## Prepare the data for model creation

```python
# Create dummy variables for columns season, mnth, weekday and weathersit

season_dum = pd.get_dummies(df.season, drop_first = True)
mnth_dum = pd.get_dummies(df.mnth, drop_first = True)
weekday_dum = pd.get_dummies(df.weekday, drop_first = True)
weathersit_dum = pd.get_dummies(df.weathersit, drop_first = True)
```

```python
# Mergin dummy variables with the data frame

df_new = pd.concat([df,season_dum,mnth_dum,weekday_dum,weathersit_dum],axis = 1)
df_new.head()
```

```python
# dropping unwanted columns 

df_new.drop(['season','mnth','weekday','weathersit'], axis = 1, inplace = True)
df_new.head()
```

```python
# Description of the new data set 

df_new.info()
```

```python
df_new.describe()
```

## Spliting the data into tranning set and test set for model creation 

```python
# Geting libaries for regression

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor 
import statsmodels.api as sm
from sklearn.metrics import r2_score
```

```python
df_train, df_test = train_test_split(df_new, train_size = 0.7, random_state = 100)
```

```python
# Shape of the data frame train set

df_train.shape
```

```python
# SHape of the data frame test set

df_test.shape
```

```python
# Using minmax scaler for scaling the variables features

scaler = MinMaxScaler()
num_col = ['temp','atemp','hum','windspeed','cnt']
df_train[num_col] = scaler.fit_transform(df_train[num_col])
df_train.head()
```

```python
# Describe the data frame train set

df_train.describe()
```

#### Here we can see that all the variables features are transform and max value is 1.

```python
# Checking variables which are hign c0-related with each other

plt.figure(figsize = (25,25))
sns.heatmap(df_train.corr(), annot = True, cmap="RdYlGn")
plt.show()
```

#### Here we can see cnt have correlation with yr variable and temp. Similarly, cloudy and hum show correlation. Spring season with Jan and Feb month, Summer season with may month and Winter season with oct and nov month show good correlation.

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Recursive feature elimination for linear model

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
# Checking VIF for the first X_train model

vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
# Dropping variables(workingday,atemp,aug,spring,dec,feb,jul,jun,mar,may,nov,oct,sat,sun,thu,wed) from X_train data set

X = X_train.drop(['workingday','atemp','aug','spring','dec','feb','jul','jun','mar','may','nov','oct','sat','sun','thu','wed'], axis = 1)
X_train_sm1 = sm.add_constant(X)
lr_1 = sm.OLS(y_train, X_train_sm1)
lr_model_1 = lr_1.fit()
lr_model_1.summary()
```

```python
# Checking VIF after dropping variables(workingday,atemp,aug,spring,dec,feb,jul,jun,mar,may,nov,oct,sat,sun,thu,wed) from X_train data set

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
# Dropping variables(hum,mon,teu,jan) from X_train data set 

X_1 = X.drop(['hum','mon','teu','jan'], axis = 1)
X_train_sm2 = sm.add_constant(X_1)
lr_2 = sm.OLS(y_train, X_train_sm2)
lr_model_2 = lr_2.fit()
lr_model_2.summary()
```

```python
# Checking VIF after dropping variables(hum,mon,teu,jan) from X_train data set

vif = pd.DataFrame()
vif['Features'] = X_1.columns
vif['VIF'] = [variance_inflation_factor(X_1.values,i) for i in range(X_1.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
# Final training model parameters

lr_model_2.params
```

```python
# Final train model columns

X_train_sm2
```

```python
y_train_pred = lr_model_2.predict(X_train_sm2)
```

```python
# Figure out the Error terms distribution 

fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)
```

#### Here we can see the curve pick is in the center(0.0) so the error terms are normaly distributed

```python
# Checking multicolinearity when VIF is less than 5

plt.figure(figsize=(15,8))
sns.heatmap(X_train_sm2.corr(),annot = True, cmap="RdYlGn")
plt.show()
```

#### Here we can see no multicolinearity is present 

```python
# Fit_transform the test model data set variables

num_col = ['temp','atemp','hum','windspeed','cnt']
df_test[num_col] = scaler.fit_transform(df_test[num_col])
df_test.head()
```

```python
# Creating X_test and y_test data sets

y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Adding constant at X_test data set

X_test_sm = sm.add_constant(X_test)
X_test_sm.head()
```

```python
# Droping unwanted variables from X_test data set

X_test_sm = X_test_sm.drop(['workingday','atemp','aug','spring','dec','feb','jul','jun','mar','may','nov','oct','sat','sun','thu','wed','hum','mon','teu','jan'], axis = 1)
X_test_sm.head()
```

```python
# R^2 value of test data set model

y_test_pred = lr_model_2.predict(X_test_sm)
r2 = r2_score(y_true = y_test, y_pred = y_test_pred)
round(r2,4)
```

```python
# Adjusted R^2 value of test data set model

adjusted_r2 = round(1-(1-r2)*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1),4)
print(adjusted_r2)
```
